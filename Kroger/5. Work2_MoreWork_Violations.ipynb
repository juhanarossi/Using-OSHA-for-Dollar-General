### 1. For Current Violations

```python
df = pd.read_csv("3_currentviolationsdrop.csv")
```

```python
df
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df.head(20)
```

```python
df.dtypes
```

```python
df['Total'] = df['Total'].astype(int)
```

```python
violations_df = df.groupby('simple_year')['Total'].sum()
```

```python
violations_df
```

```python
new2_df = violations_df.to_frame(name='violations')
```

```python
new2_df
```

```python
new2_df.to_csv('5_CurrentViolations.csv')
```

### 2. For Initial Violations

```python
df = pd.read_csv("3_initialviolationsdrop.csv")
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df['Total'] = df['Total'].astype(int)
```

```python
violations_df = df.groupby('simple_year')['Total'].sum()
```

```python
violations_df
```

```python
new2_df = violations_df.to_frame(name='violations')
```

```python
new2_df
```

```python
new2_df.to_csv('5_InitialViolations.csv')
```
