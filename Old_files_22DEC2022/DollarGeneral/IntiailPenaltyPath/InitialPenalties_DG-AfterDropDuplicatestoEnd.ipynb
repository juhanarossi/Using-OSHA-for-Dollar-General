```python
df = pd.read_csv("Dec22DGINITIAL_toDropDuplicates.csv")
```

```python
df
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df.Total = df.Total.astype(str)
```

```python
df.Total = df.Total.str.replace("$", "")
```

```python
df.Total = df.Total.str.replace(",", "")
```

```python
df.Total = df.Total.astype(int)
```

```python
dgfines_df = df.groupby('simple_year')['Total'].sum()
```

```python
dgfines_df
```

```python
new_df = dgfines_df.to_frame(name='InitialFines')
```

```python
new_df
```

```python
new_df.to_csv('Dec22DG_Final_InitialPenalty.csv', index=False)
```
