```python
df = pd.read_csv("2_final_df.csv")
```

```python
df
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df.head(20)
```

```python
df.dtypes
```

### 1. For Current Penalty

```python
Current_Penalty_df = df[df['Violations/Penalties'] == 'Current Penalty']
```

```python
Current_Penalty_df.sample(20)
```

```python
Current_Penalty_df['Total']=Current_Penalty_df['Total'].str.replace(',','')
```

```python
Current_Penalty_df['Total']=Current_Penalty_df['Total'].str.replace('$','')
```

```python
Current_Penalty_df['Total'] = Current_Penalty_df['Total'].astype(int)
```

```python
Current_Penalty_df
```

```python
fines_df = Current_Penalty_df.groupby('simple_year')['Total'].sum()
```

```python
fines_df
```

```python
new_df = fines_df.to_frame(name='fines')
```

```python
new_df
```

```python
new_df.to_csv('3_CurrentFines.csv')
```
