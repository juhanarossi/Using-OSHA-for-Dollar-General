### 1. For Current Penalty

```python
df = pd.read_csv("3_currentpenaltydrop.csv")
```

```python
df
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df.head(20)
```

```python
df['Total']=df['Total'].str.replace(',','')
```

```python
df['Total']=df['Total'].str.replace('$','')
```

```python
df['Total'] = df['Total'].astype(int)
```

```python
df
```

```python
current_fines_df = df.groupby('simple_year')['Total'].sum()
```

```python
current_fines_df
```

```python
new_df = current_fines_df.to_frame(name='fines')
```

```python
new_df
```

```python
new_df.to_csv('4_CurrentFines.csv')
```
