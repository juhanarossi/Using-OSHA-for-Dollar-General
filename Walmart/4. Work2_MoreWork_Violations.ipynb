```python
df = pd.read_csv("2_final_df.csv")
```

```python
df
```

```python
df['year'] = pd.to_datetime(df['Open Date'], format='%m/%d/%Y')
```

```python
df['simple_year'] = df['year'].dt.year
```

```python
df.head(20)
```

```python
df.dtypes
```

### 1. For Current Violations

```python
Current_Violation_df = df[df['Violations/Penalties'] == 'Current Violations']
```

```python
Current_Violation_df.sample(20)
```

```python
Current_Violation_df['Total'] = Current_Violation_df['Total'].astype(int)
```

```python
violations_df = Current_Violation_df.groupby('simple_year')['Total'].sum()
```

```python
violations_df
```

```python
new2_df = violations_df.to_frame(name='violations')
```

```python
new2_df
```

```python
new2_df.to_csv('4_CurrentViolations.csv')
```

### 2. For Initial Violations

```python
Initial_Violation_df = df[df['Violations/Penalties'] == 'Initial Violations']
```

```python
Initial_Violation_df['Total'] = Initial_Violation_df['Total'].astype(int)
```

```python
violations_df = Initial_Violation_df.groupby('simple_year')['Total'].sum()
```

```python
violations_df
```

```python
new2_df = violations_df.to_frame(name='violations')
```

```python
new2_df
```

```python
new2_df.to_csv('4_InitialViolations.csv')
```
