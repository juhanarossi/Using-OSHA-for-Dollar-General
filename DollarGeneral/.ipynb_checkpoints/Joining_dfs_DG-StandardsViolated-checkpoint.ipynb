```python
import pandas as pd
pd.options.display.max_colwidth = None
pd.options.display.max_columns = None
```

```python
df1 = pd.read_csv("Dec22DG_activities_df.csv")
df1
```

```python
df2 = pd.read_csv("Dec22DG_details_df.csv")
df2
```

```python
df3 = pd.read_csv("Dec22DG_violation_items_df.csv")
df3
```

```python
master_df = pd.merge(df1, df2, how='inner', left_on='Activity', right_on='Inspection Nr')
```

```python
master_df
```

```python
final_df = pd.merge(master_df, df3, how='inner', left_on='Activity', right_on='activity_code')
```

```python
final_df
```

```python
final_df.to_csv('Dec22DG_TOTAL_VIOLATIONS_master.csv')
```
