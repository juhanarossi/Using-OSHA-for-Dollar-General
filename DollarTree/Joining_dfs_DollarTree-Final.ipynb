```python
import pandas as pd
pd.options.display.max_colwidth = None
pd.options.display.max_columns = None
```

```python
df1 = pd.read_csv("Dec22DT_activities_df.csv")
```

```python
df2 = pd.read_csv("Dec22DT_details_df.csv")
```

```python
df3 = pd.read_csv("Dec22DT_violation_summary_df.csv")
df3
```

```python
df3 = df3[df3['Violations/Penalties'] == 'Current Penalty']
df3
```

```python
master_df = pd.merge(df1, df2, how='inner', left_on='Activity', right_on='Inspection Nr')
```

```python
master_df
```

```python
final_df = pd.merge(master_df, df3, how='inner', left_on='Activity', right_on='activity_code')
```

```python
final_df
```

```python
final_df.to_csv('Dec22DT_TOTAL_CurrentPenalty_master.csv')
```
